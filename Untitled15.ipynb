```python
import time
import cv2
import os

def video_to_frame(input_loc, output_loc):
    
    
    try:
        os.mkdir(output_loc)
    except OSError:
        pass
    time_start = time.time()
    cap = cv2.VideoCapture(input_loc)
    video_length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) - 1

    print ("Number of Frames: ", video_length)
    count = 0
    print ("Converting video...\n")
    while cap.isOpened():
        ret, frame = cap.read()
        cv2.imwrite(output_loc + "051(%d).jpg" % (count+1), frame)
        count = count + 1
        if (count > (video_length-1)):
            time_end = time.time()
            cap.release()
            print ("Done extracting frame. \n%d frames extracted" %count)
            print ("It took %d seconds for conversion." %(time_end-time_start))
            break
    
    
        
input_loc = r'C:\Users\DELL\Documents\sample1\video\kbm.mp4'
#output_loc = r'C:\Users\DELL\output\frames'
output_loc = r'C:\Users\DELL\Documents\sample\ '
video_to_frame(input_loc, output_loc)
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
for file in os.listdir(r'C:\Users\DELL\Documents\sample'):
    print(file)
    full_path = 'C:/Users/DELL/Documents/sample/' + file
    
    print(full_path)
    
    image = load_img(full_path, target_size=(224,224))
    image = img_to_array(image)
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    image = preprocess_input(image)
    y_pred = model.predict(image)
    label = decode_predictions(y_pred, top =1)
    print(label)
    print()
```
